```python
hn_karma_clique_nodes = [e for e in hn_interaction_graph.nodes(data=True)
                      if len(hn_interaction_graph.neighbors(e[0])) > 4]
su_karma_clique_nodes = [e for e in su_interaction_graph.nodes(data=True) 
                         if len(su_interaction_graph.neighbors(e[0])) > 4]
hn_karmas = [x[1].get('karma', 0) for x in hn_karma_clique_nodes]
su_karmas = [x[1].get('reputation', 0) for x in su_karma_clique_nodes]
```

```python
def avg_avg(karmas, neighbor):
    return [(k, np.mean(list(e[1] for e in g))) for k, g in itertools.groupby(
                            sorted(e for e in itertools.izip(karmas, neighbor)),
                                               key=lambda x: x[0])]

hn_outbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=True) or [0])
                              for n in hn_karma_clique_nodes]
hn_outbound_avg_avg = avg_avg(hn_karmas, hn_outbound_neighbor_karma)
hn_outbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=True) or [0])
                              for n in hn_karma_clique_nodes]
hn_inbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=False) or [0])
                              for n in hn_karma_clique_nodes]
hn_inbound_avg_avg = avg_avg(hn_karmas, hn_inbound_neighbor_karma)
hn_inbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=False) or [0])
                              for n in hn_karma_clique_nodes]
su_outbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=True) or [0])
                              for n in su_karma_clique_nodes]
su_outbound_avg_avg = avg_avg(su_karmas, su_outbound_neighbor_karma)
su_outbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=True) or [0])
                              for n in su_karma_clique_nodes]
su_inbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=False) or [0])
                              for n in su_karma_clique_nodes]
su_inbound_avg_avg = avg_avg(su_karmas, su_inbound_neighbor_karma)
su_inbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=False) or [0])
                              for n in su_karma_clique_nodes]
```

Interpretation:

The above plots show the average karma/reputation of an anchor node's neighbors, plotted against the anchor's karma. For Hacker News, this we see an enveloped shape of data encircling both the outbound karma and the inbound karma data points. While the variance decreases with larger anchor karma, the mean appears to be approximately the same regardless of anchor karma - somewhere near 1000 or so. Since this doesn't seem to vary with the anchors karma and is relatively large, it suggests that there is a relatively heterogenous group of commenters with mid-level karma. Instead of "karma cliques", we instead find "major posters" that seem to reply all the way across of the spectrum.

Super User's plots are dramatically different than Hacker News. While there is no discernable pattern based on the anchor node's karma, there is a noticable clustering between looking at inbound reputation and outbound reputation. We see that the inbound reputation (people answering this post) tends to be much higher than the outbound karma (the questioner's karma). This suggests two distinct roles on SuperUser: the questioners and the answers.

```python
def pickle2feature_mapping(graph, mapping_file, pickle_file, default=0.0):
    with open(mapping_file, "rb") as f:
        u2id = {v:k for k, v in pickle.load(f).iteritems()}
    with open(pickle_file, "rb") as f:
        d = pickle.load(f)
    return [d.get(u2id[u], default) for u in graph.nodes()]

def pickle2feature(graph, pickle_file, default=0.0):
    with open(pickle_file, "rb") as f:
        d = pickle.load(f)
    return [d.get(u, default) for u in graph.nodes()]

hn_karmas = [x[1].get('karma', 0) for x in hn_interaction_graph.nodes(data=True)]
hn_num_posts = [hn_interaction_graph.out_degree(x, weight='weight') for x in hn_interaction_graph.nodes()]
hn_seconds_since_join = [(dt.datetime.now() - x[1].get('join_date', dt.datetime.now())).total_seconds() 
                         for x in hn_interaction_graph.nodes(data=True)]
hn_in_degrees = [hn_interaction_graph.in_degree(x) for x in hn_interaction_graph.nodes()]
assert(len(hn_karmas) == len(hn_num_posts))
assert(len(hn_karmas) == len(hn_seconds_since_join))
hn_pagerank_dict = graph_tools.pagerank(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_pagerank")
hn_pageranks = [hn_pagerank_dict[x] for x in hn_interaction_graph.nodes()]
del hn_pagerank_dict
hn_weighted_pagerank_dict = graph_tools.weighted_pagerank(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_weighted_pagerank")
hn_weighted_pageranks = [hn_weighted_pagerank_dict[x] for x in hn_interaction_graph.nodes()]
del hn_weighted_pagerank_dict
hn_hubs_dict, hn_authorities_dict = graph_tools.hits(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_hits")
hn_hubs = [hn_hubs_dict[x] for x in hn_interaction_graph.nodes()]
hn_authorities = [hn_authorities_dict[x] for x in hn_interaction_graph.nodes()]
del hn_hubs_dict
del hn_authorities_dict

hn_constraint = pickle2feature_mapping(hn_interaction_graph, "../data/hn_mapping.pickle", "../data/hn_constraint.pickle")
hn_betweenness = pickle2feature_mapping(hn_interaction_graph, "../data/hn_mapping.pickle", "../data/hn_betweenness.pickle")
hn_closeness = pickle2feature_mapping(hn_interaction_graph, "../data/hn_mapping.pickle", "../data/hn_closeness.pickle")

su_reputations = [x[1]["reputation"] for x in su_interaction_graph.nodes(data=True)]
su_num_posts = [su_interaction_graph.out_degree(x, weight='weight') for x in su_interaction_graph.nodes()]
su_seconds_since_join = [(dt.datetime.now() - x[1].get('creation_time', dt.datetime.now())).total_seconds() 
                         for x in su_interaction_graph.nodes(data=True)]
su_in_degrees = [su_interaction_graph.in_degree(x) for x in su_interaction_graph.nodes()]

assert(len(su_reputations) == len(su_seconds_since_join))
su_pagerank_dict = graph_tools.pagerank(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_pagerank")
su_pageranks = [su_pagerank_dict[x] for x in su_interaction_graph.nodes()]
del su_pagerank_dict
su_weighted_pagerank_dict = graph_tools.weighted_pagerank(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_weighted_pagerank")
su_weighted_pageranks = [su_weighted_pagerank_dict[x] for x in su_interaction_graph.nodes()]
del su_weighted_pagerank_dict

su_hubs_dict, su_authorities_dict = graph_tools.hits(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_hits")
su_hubs = [su_hubs_dict[x] for x in su_interaction_graph.nodes()]
su_authorities = [su_authorities_dict[x] for x in su_interaction_graph.nodes()]
del su_hubs_dict
del su_authorities_dict

su_constraint = pickle2feature(su_interaction_graph, "../data/su_constraint.pickle")
su_betweenness = pickle2feature(su_interaction_graph, "../data/su_betweenness.pickle")
su_closeness = pickle2feature(su_interaction_graph, "../data/su_closeness.pickle")
```

```python
# LDA
HN_LDA_USER_FILE = "../data/lda/hn.user_lda.pickle"
HN_LDA_PR_FILE = "../data/lda/hn.lda_ranks.pickle"
SU_LDA_USER_FILE = "../data/lda/su.user_lda.pickle"
SU_LDA_PR_FILE = "../data/lda/su.lda_ranks.pickle"

def read_ldas(graph, user_file, pr_file):
    with open(pr_file) as f:
        lda_pr_dict = pickle.load(f)
    
    with open(user_file) as f:
        lda_dict = pickle.load(f)
        
    lda_pr = np.array([lda_pr_dict[user] for user in graph.nodes()])
    lda = np.zeros((graph.number_of_nodes(), 10))
    for i, user in enumerate(graph.nodes()):
        if user in lda_dict:
            for j, val in lda_dict[user]:
                lda[i, j] = val
    
    return lda, lda_pr

hn_lda, hn_lda_pr = read_ldas(hn_interaction_graph, HN_LDA_USER_FILE, HN_LDA_PR_FILE)
su_lda, su_lda_pr = read_ldas(su_interaction_graph, SU_LDA_USER_FILE, SU_LDA_PR_FILE)
```

#### Results

```python
def print_performance(model, input, gold, prediction, title=""):
    print "\t%s" % title
    print "\t\tRoot mean square error: %f" % np.sqrt(np.mean((prediction - gold)**2))
    print "\t\tCoefficient of determination: %f" % model.score(input, gold)
    for start, end in [(1,10), (11, 100), (101, 1000), (1001, 10000), (10001, 100000)]:
        l = (gold >= start) & (gold <= end)
        m = np.sqrt(np.mean((prediction[l] - gold[l])**2))
        print "\t\t\t[%s,%s] Root mean square error: %f" % (start, end, m)
        print "\t\t\t[%s,%s] Normalized root mean square error: %f" % (start, end, m / (end - start))
        
    

predict_hn_validate = hn_regr.predict(hn_X_validate)
predict_hn_tree_validate = hn_tree_regression.predict(hn_X_validate)
predict_hn_forest_validate = hn_forest_regression.predict(hn_X_validate)

print "Hacker news summary:"
print_performance(hn_regr, hn_X_validate, hn_y_validate, predict_hn_validate,
                  title="Linear Regression")
print_performance(hn_tree_regression, hn_X_validate, hn_y_validate, predict_hn_tree_validate, 
                  title="Decision Tree Regression")
print_performance(hn_forest_regression, hn_X_validate, hn_y_validate, predict_hn_forest_validate, 
                  title="Random Forest Regression")


predict_su_validate = su_regr.predict(su_X_validate)
predict_su_tree_validate = su_tree_regression.predict(su_X_validate)
predict_su_forest_validate = su_forest_regression.predict(su_X_validate)

print "Super User summary:"
print_performance(su_regr, su_X_validate, su_y_validate, predict_su_validate,
                  title="Linear Regression")
print_performance(su_tree_regression, su_X_validate, su_y_validate, predict_su_tree_validate, 
                  title="Decision Tree Regression")
print_performance(su_forest_regression, su_X_validate, su_y_validate, predict_su_forest_validate, 
                  title="Random Forest Regression")
```
