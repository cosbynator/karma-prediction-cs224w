#### Karma Cliques

```python
hn_karma_clique_nodes = [e for e in hn_interaction_graph.nodes(data=True)
                      if len(hn_interaction_graph.neighbors(e[0])) > 4]
su_karma_clique_nodes = [e for e in su_interaction_graph.nodes(data=True) 
                         if len(su_interaction_graph.neighbors(e[0])) > 4]
hn_karmas = [x[1].get('karma', 0) for x in hn_karma_clique_nodes]
su_karmas = [x[1].get('reputation', 0) for x in su_karma_clique_nodes]
```

```python
def avg_avg(karmas, neighbor):
    return [(k, np.mean(list(e[1] for e in g))) for k, g in itertools.groupby(
                            sorted(e for e in itertools.izip(karmas, neighbor)),
                                               key=lambda x: x[0])]

hn_outbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=True) or [0])
                              for n in hn_karma_clique_nodes]
hn_outbound_avg_avg = avg_avg(hn_karmas, hn_outbound_neighbor_karma)
hn_outbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=True) or [0])
                              for n in hn_karma_clique_nodes]
hn_inbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=False) or [0])
                              for n in hn_karma_clique_nodes]
hn_inbound_avg_avg = avg_avg(hn_karmas, hn_inbound_neighbor_karma)
hn_inbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(hn_interaction_graph, n[0], outbound=False) or [0])
                              for n in hn_karma_clique_nodes]
su_outbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=True) or [0])
                              for n in su_karma_clique_nodes]
su_outbound_avg_avg = avg_avg(su_karmas, su_outbound_neighbor_karma)
su_outbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=True) or [0])
                              for n in su_karma_clique_nodes]
su_inbound_neighbor_karma = [np.mean(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=False) or [0])
                              for n in su_karma_clique_nodes]
su_inbound_avg_avg = avg_avg(su_karmas, su_inbound_neighbor_karma)
su_inbound_neighbor_karma_std = [np.std(graph_tools.neighborhood_karma(su_interaction_graph, n[0], outbound=False) or [0])
                              for n in su_karma_clique_nodes]
```

```python
fig = plt.figure(figsize=(24,24))
ax = fig.add_subplot(221)
ax.set_title("Hacker News Neighbor Karma Distribution")
ax.set_xlabel("Karma")
ax.set_ylabel("Neighbor Karma")
ax.set_yscale('log')
ax.set_xscale('log')
ax.scatter(hn_karmas, hn_inbound_neighbor_karma, marker='x', color='b', label="Avg Inbound Neighbor Karma")
ax.scatter(hn_karmas, hn_outbound_neighbor_karma, marker='+', color='g', label="Avg Outbound Neighbor Karma")

ax.legend(loc="lower left")

ax = fig.add_subplot(223)
ax.set_title("Hacker News Neighbor Karma Line")
ax.set_xlabel("Karma")
ax.set_ylabel("Neighbor Karma Average of Average")
ax.set_yscale('log')
ax.set_xscale('log')
ax.scatter(hn_karmas, hn_inbound_neighbor_karma, marker='x', color='b', label=None)
ax.scatter(hn_karmas, hn_outbound_neighbor_karma, marker='+', color='g', label=None)
ax.plot([e[0] for e in hn_outbound_avg_avg], 
       [e[1] for e in hn_outbound_avg_avg], color='r', ls="-", label='Avg of Avg Outbound')
ax.plot([e[0] for e in hn_inbound_avg_avg], 
       [e[1] for e in hn_inbound_avg_avg], color='k', ls=":", label='Avg of Avg Inbound')
ax.legend(loc="lower left")

ax = fig.add_subplot(222)
ax.set_title("Super User Neighbor Karma Distribution")
ax.set_xlabel("Reputation")
ax.set_ylabel("Avg Neighbor Reputation")
ax.set_yscale('log')
ax.set_xscale('log')
ax.scatter(su_karmas, su_inbound_neighbor_karma, marker='x', color='r', label="Avg Inbound Neighbor Reputation")
ax.scatter(su_karmas, su_outbound_neighbor_karma, marker='+', color='y', label="Avg Outbound Neighbor Reputation")

ax.legend(loc="lower left")

ax = fig.add_subplot(224)
ax.set_title("Super User Neighbor Karma Line")
ax.set_xlabel("Reputation")
ax.set_ylabel("Neighbor Reputation Average of Average")
ax.set_yscale('log')
ax.set_xscale('log')
ax.scatter(su_karmas, su_inbound_neighbor_karma, marker='x', color='r', label=None)
ax.scatter(su_karmas, su_outbound_neighbor_karma, marker='+', color='y', label=None)
ax.plot([e[0] for e in su_inbound_avg_avg], 
       [e[1] for e in su_inbound_avg_avg], color='b', ls="-", label='Avg of Avg Inbound')
ax.plot([e[0] for e in su_outbound_avg_avg], 
       [e[1] for e in su_outbound_avg_avg], color='k', ls=":", label='Avg of Avg Outbound')
ax.legend(loc="lower left")
```

Interpretation:

The above plots show the average karma/reputation of an anchor node's neighbors, plotted against the anchor's karma. For Hacker News, this we see an enveloped shape of data encircling both the outbound karma and the inbound karma data points. While the variance decreases with larger anchor karma, the mean appears to be approximately the same regardless of anchor karma - somewhere near 1000 or so. Since this doesn't seem to vary with the anchors karma and is relatively large, it suggests that there is a relatively heterogenous group of commenters with mid-level karma. Instead of "karma cliques", we instead find "major posters" that seem to reply all the way across of the spectrum.

Super User's plots are dramatically different than Hacker News. While there is no discernable pattern based on the anchor node's karma, there is a noticable clustering between looking at inbound reputation and outbound reputation. We see that the inbound reputation (people answering this post) tends to be much higher than the outbound karma (the questioner's karma). This suggests two distinct roles on SuperUser: the questioners and the answers.

### Build features

```python
hn_karmas = [x[1].get('karma', 0) for x in hn_interaction_graph.nodes(data=True)]
hn_num_posts = [hn_interaction_graph.out_degree(x, weight='weight') for x in hn_interaction_graph.nodes()]
hn_seconds_since_join = [(dt.datetime.now() - x[1].get('join_date', dt.datetime.now())).total_seconds() 
                         for x in hn_interaction_graph.nodes(data=True)]
assert(len(hn_karmas) == len(hn_num_posts))
assert(len(hn_karmas) == len(hn_seconds_since_join))
hn_pagerank_dict = graph_tools.pagerank(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_pagerank")
hn_pageranks = [hn_pagerank_dict[x] for x in hn_interaction_graph.nodes()]
del hn_pagerank_dict
hn_weighted_pagerank_dict = graph_tools.weighted_pagerank(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_weighted_pagerank")
hn_weighted_pageranks = [hn_weighted_pagerank_dict[x] for x in hn_interaction_graph.nodes()]
del hn_weighted_pagerank_dict
hn_hubs_dict, hn_authorities_dict = graph_tools.hits(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_hits")
hn_hubs = [hn_hubs_dict[x] for x in hn_interaction_graph.nodes()]
hn_authorities = [hn_authorities_dict[x] for x in hn_interaction_graph.nodes()]
del hn_hubs_dict
del hn_authorities_dict

hn_feature_matrix = np.vstack((hn_num_posts, hn_seconds_since_join, hn_weighted_pageranks, 
                               hn_hubs, hn_authorities,
                               np.ones(len(hn_num_posts)))).T

su_reputations = [x[1]["reputation"] for x in su_interaction_graph.nodes(data=True)]
su_num_posts = [su_interaction_graph.out_degree(x, weight='weight') for x in su_interaction_graph.nodes()]
su_seconds_since_join = [(dt.datetime.now() - x[1].get('creation_time', dt.datetime.now())).total_seconds() 
                         for x in su_interaction_graph.nodes(data=True)]
assert(len(su_reputations) == len(su_seconds_since_join))
su_pagerank_dict = graph_tools.pagerank(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_pagerank")
su_pageranks = [su_pagerank_dict[x] for x in su_interaction_graph.nodes()]
del su_pagerank_dict
su_weighted_pagerank_dict = graph_tools.weighted_pagerank(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_weighted_pagerank")
su_weighted_pageranks = [su_weighted_pagerank_dict[x] for x in su_interaction_graph.nodes()]
del su_weighted_pagerank_dict

su_hubs_dict, su_authorities_dict = graph_tools.hits(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_hits")
su_hubs = [su_hubs_dict[x] for x in su_interaction_graph.nodes()]
su_authorities = [su_authorities_dict[x] for x in su_interaction_graph.nodes()]
del su_hubs_dict
del su_authorities_dict

su_feature_matrix = np.vstack((su_num_posts, su_seconds_since_join, su_weighted_pageranks,
                               su_hubs, su_authorities,
                               np.ones(len(su_num_posts)))).T
```

## Learning

### Split data

```python
RANDOM_SEED = 2358289 # reusable seed
hn_X, hn_y = shuffle(hn_feature_matrix, hn_karmas, random_state=RANDOM_SEED)
su_X, su_y = shuffle(su_feature_matrix, su_reputations, random_state=RANDOM_SEED)
```

```python
TRAIN_PERCENT = 0.7
VALIDATE_PERCENT = 0.15
hn_n_samples = hn_feature_matrix.shape[0]
su_n_samples = su_feature_matrix.shape[0]
hn_train_limit = int(TRAIN_PERCENT * hn_n_samples)
hn_validate_limit = int((TRAIN_PERCENT + VALIDATE_PERCENT) * hn_n_samples)
su_validate_limit = int((TRAIN_PERCENT + VALIDATE_PERCENT) * su_n_samples)
su_train_limit = int(TRAIN_PERCENT * su_n_samples)
hn_X_train, hn_X_validate, hn_X_test = hn_X[:hn_train_limit], hn_X[hn_train_limit:hn_validate_limit], hn_X[hn_validate_limit:]
su_X_train, su_X_validate, su_X_test = su_X[:su_train_limit], su_X[su_train_limit:su_validate_limit], su_X[su_validate_limit:]
hn_y_train, hn_y_validate, hn_y_test = hn_y[:hn_train_limit], hn_y[hn_train_limit:hn_validate_limit], hn_y[hn_validate_limit:]
su_y_train, su_y_validate, su_y_test = su_y[:su_train_limit], su_y[su_train_limit:su_validate_limit], su_y[su_validate_limit:]
```

### Regression

```python
hn_regr = linear_model.LinearRegression()
hn_regr.fit(hn_X_train, hn_y_train)
```

```python
su_regr = linear_model.LinearRegression()
su_regr.fit(su_X_train, su_y_train)
```

```python
# Test linear regression

predict_hn_validate = hn_regr.predict(hn_X_validate)
predict_su_validate = su_regr.predict(su_X_validate)

print "Hacker news summary:"
print "- Root mean square error: %f" % np.sqrt(np.mean((predict_hn_validate - hn_y_validate)**2))
print "- Coefficient of determination: %f" % hn_regr.score(hn_X_validate, hn_y_validate)

print "Super User summary:"
print "- Root mean square error: %f" % np.sqrt(np.mean((predict_su_validate - su_y_validate)**2))
print "- Coefficient of determination: %f" % su_regr.score(su_X_validate, su_y_validate)
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(121)
ax.set_xlabel("Predicted Karma")
ax.set_ylabel("Actual Karma")
ax.scatter(predict_hn_validate, hn_y_validate)
ax.set_title("HN Predicted Vs. Actual")

ax = fig.add_subplot(122)
ax.set_title("SuperUser Predicted vs. Actual")
ax.set_xlabel("Predicted Karma")
ax.set_ylabel("Actual Karma")
ax.scatter(predict_su_validate, su_y_validate)
```

### Classification

```python
HN_FAMOUS_THRESHOLD = 2500
SU_FAMOUS_THRESHOLD = 1000
```

```python
hn_is_famous_train = hn_y_train > HN_FAMOUS_THRESHOLD
su_is_famous_train = su_y_train > SU_FAMOUS_THRESHOLD
print "HN number of famous nodes (train): %d" % np.sum(hn_is_famous_train)
print "HN number of non-famous nodes (train): %d" % np.sum(np.invert(hn_is_famous_train))
print "SU number of famous nodes (train): %d" % np.sum(su_is_famous_train)
print "SU number of non-famous nodes (train): %d" % np.sum(np.invert(su_is_famous_train))
```

```python
hn_is_famous_validate = hn_y_validate > HN_FAMOUS_THRESHOLD
su_is_famous_validate = su_y_validate > SU_FAMOUS_THRESHOLD
print "HN number of famous nodes (validation): %d" % np.sum(hn_is_famous_validate)
print "HN number of non-famous nodes (validation): %d" % np.sum(np.invert(hn_is_famous_validate))
print "SU number of famous nodes (validation): %d" % np.sum(su_is_famous_validate)
print "SU number of non-famous nodes (validation): %d" % np.sum(np.invert(su_is_famous_validate))
```

```python
hn_logistic = linear_model.LogisticRegression(C=1e5)
hn_probs = hn_logistic.fit(hn_X_train, hn_is_famous_train).predict_proba(hn_X_validate)
```

```python
su_logistic = linear_model.LogisticRegression(C=1e5)
su_probs = su_logistic.fit(su_X_train, su_is_famous_train).predict_proba(su_X_validate)
```

```python
hn_fpr, hn_tpr, _ = roc_curve(hn_is_famous_validate, hn_probs[:, 1])
su_fpr, su_tpr, _ = roc_curve(su_is_famous_validate, su_probs[:, 1])
hn_roc_auc = auc(hn_fpr, hn_tpr)
su_roc_auc = auc(su_fpr, su_tpr)

plt.figure(figsize=(8,8))
plt.plot(hn_fpr, hn_tpr, color='b', linewidth=2, label='Hacker News ROC curve (area = %0.2f)' % hn_roc_auc)
plt.plot(su_fpr, su_tpr, color='r', linewidth=2, label='Super User ROC curve (area = %0.2f)' % su_roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.title("Receiver Operating Characteristic")
plt.legend(loc="lower right")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.savefig(os.path.join(FIGURES_DIR, "classification_roc.pdf"))
```

```python
hn_precision, hn_recall, _ = precision_recall_curve(hn_is_famous_validate, hn_probs[:, 1])
su_precision, su_recall, _ = precision_recall_curve(su_is_famous_validate, su_probs[:, 1])

plt.figure(figsize=(8,8))
plt.plot(hn_recall, hn_precision, color='b', linewidth=2, label='Hacker News Precision-Recall curve')
plt.plot(su_recall, su_precision, color='r', linewidth=2, label='Super User Precision-Recall curve')
plt.plot([0, 1], [1, 0], 'k--')
plt.title('Precision-Recall')
plt.legend(loc="lower left")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.savefig(os.path.join(FIGURES_DIR, "classification_pr_curve.pdf"))
```
