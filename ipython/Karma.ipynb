## Setup

```python
%matplotlib inline
%load_ext autoreload
%autoreload 1

import os
import sys

sys.path.append("../src/python")

import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import powerlaw
from sklearn import linear_model
from sklearn.utils import shuffle
from sklearn.metrics import roc_curve, precision_recall_curve, auc

from util import xs, ys, mkcdf, mkccdf, mkpdf
%aimport hn_karma
%aimport su_reputation
%aimport graph_tools

# Data files
HN_DB_FILE = "../data/hn.db"
SU_DB_FILE = "../data/superuser.db"
CACHE_DIR = "../data/cache"

FIGURES_DIR = "../reports/figures"
```

## Data Analysis

### Build graphs

```python
hn_interaction_graph = hn_karma.nx_interaction_graph(HN_DB_FILE, cache_dir=CACHE_DIR)
print "Number of edges: %d" % hn_interaction_graph.number_of_edges()
print "Number of nodes: %d" % hn_interaction_graph.number_of_nodes()
```

```python
su_interaction_graph = su_reputation.nx_interaction_graph(SU_DB_FILE, cache_dir=CACHE_DIR)
print "Number of edges: %d" % su_interaction_graph.number_of_edges()
print "Number of nodes: %d" % su_interaction_graph.number_of_nodes()
```

### Statistics

```python
hn_stats = graph_tools.compute_network_stats(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_stats")
su_stats = graph_tools.compute_network_stats(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_stats")
print graph_tools.network_stats_table(hn_stats, su_stats)
```

### Build features

```python
hn_karmas = [x[1].get('karma', 0) for x in hn_interaction_graph.nodes(data=True)]
hn_num_posts = [hn_interaction_graph.out_degree(x, weight='weight') for x in hn_interaction_graph.nodes()]
hn_seconds_since_join = [(dt.datetime.now() - x[1].get('join_date', dt.datetime.now())).total_seconds() 
                         for x in hn_interaction_graph.nodes(data=True)]
assert(len(hn_karmas) == len(hn_num_posts))
assert(len(hn_karmas) == len(hn_seconds_since_join))
hn_pagerank_dict = graph_tools.pagerank(hn_interaction_graph, cache_dir=CACHE_DIR, cache_key="hn_pagerank")
hn_pageranks = [hn_pagerank_dict[x] for x in hn_interaction_graph.nodes()]
del hn_pagerank_dict

hn_feature_matrix = np.vstack((hn_num_posts, hn_seconds_since_join, np.ones(len(hn_num_posts)))).T

su_reputations = [x[1]["reputation"] for x in su_interaction_graph.nodes(data=True)]
su_num_posts = [su_interaction_graph.out_degree(x, weight='weight') for x in su_interaction_graph.nodes()]
su_seconds_since_join = [(dt.datetime.now() - x[1].get('creation_time', dt.datetime.now())).total_seconds() 
                         for x in su_interaction_graph.nodes(data=True)]
su_pagerank_dict = graph_tools.pagerank(su_interaction_graph, cache_dir=CACHE_DIR, cache_key="su_pagerank")
su_pageranks = [su_pagerank_dict[x] for x in su_interaction_graph.nodes()]
del su_pagerank_dict
assert(len(su_reputations) == len(su_num_posts))
assert(len(su_reputations) == len(su_seconds_since_join))
su_feature_matrix = np.vstack((su_num_posts, su_seconds_since_join, np.ones(len(su_num_posts)))).T
```

### Fit powerlaws

```python
fit = powerlaw.Fit([e for e in hn_karmas if e > 0])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
fit.plot_ccdf(color='b', linewidth=2, label="Karma CCDF", ax=ax)
fit.power_law.plot_ccdf(color='b', linestyle='--', ax=ax,
                        label=r"Power Law $\alpha=%.1f,x_{min}=%.1f$" % (fit.power_law.alpha, fit.power_law.xmin))
fit.truncated_power_law.plot_ccdf(color='b', linestyle=':', ax=ax,
                        label=r"Damped Power law $\alpha=%.1f,x_{min}=%.1f,\lambda=%.2E$" % (fit.truncated_power_law.alpha, 
                                                                                    fit.truncated_power_law.xmin,
                                                                                    fit.truncated_power_law.Lambda))
ax.set_title("Hacker News Karma Distribution")
ax.set_xlabel("Karma")
ax.set_ylabel("Frequency")
ax.legend(loc="lower left")
fig.savefig(os.path.join(FIGURES_DIR, "hn_karma_distribution.pdf"))
```

```python
fit = powerlaw.Fit([max(e - 100, 1) for e in su_reputations if e > 0])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
fit.plot_ccdf(color='r', linewidth=2, label="Karma CCDF", ax=ax)
fit.power_law.plot_ccdf(color='r', linestyle='--', ax=ax,
                        label=r"Power Law $\alpha=%.1f,x_{min}=%.1f$" % (fit.power_law.alpha, fit.power_law.xmin))
ax.set_title("Superuser Reputation Distribution")
ax.set_xlabel("Reputation")
ax.set_ylabel("Frequency")
ax.set_xlim(xmin=1)
ax.legend(loc="lower left")
fig.savefig(os.path.join(FIGURES_DIR, "su_karma_distribution.pdf"))
```

## Learning

### Split data

```python
RANDOM_SEED = 2358289 # reusable seed
hn_X, hn_y = shuffle(hn_feature_matrix, hn_karmas, random_state=RANDOM_SEED)
su_X, su_y = shuffle(su_feature_matrix, su_reputations, random_state=RANDOM_SEED)
```

```python
TRAIN_PERCENT = 0.7
hn_n_samples = hn_feature_matrix.shape[0]
su_n_samples = su_feature_matrix.shape[0]
hn_train_limit = int(TRAIN_PERCENT * hn_n_samples)
su_train_limit = int(TRAIN_PERCENT * su_n_samples)
hn_X_train, hn_X_test = hn_X[:hn_train_limit], hn_X[hn_train_limit:]
su_X_train, su_X_test = su_X[:su_train_limit], su_X[su_train_limit:]
hn_y_train, hn_y_test = hn_y[:hn_train_limit], hn_y[hn_train_limit:]
su_y_train, su_y_test = su_y[:su_train_limit], su_y[su_train_limit:]
```

### Regression

```python
hn_regr = linear_model.LinearRegression()
hn_regr.fit(hn_X_train, hn_y_train)
```

```python
su_regr = linear_model.LinearRegression()
su_regr.fit(su_X_train, su_y_train)
```

```python
# Test linear regression

print "Hacker news summary:"
print "- Root mean square error: %f" % np.sqrt(np.mean((hn_regr.predict(hn_X_test) - hn_y_test)**2))
print "- Coefficient of determination: %f" % hn_regr.score(hn_X_test, hn_y_test)

print "Super User summary:"
print "- Root mean square error: %f" % np.sqrt(np.mean((su_regr.predict(su_X_test) - su_y_test)**2))
print "- Coefficient of determination: %f" % su_regr.score(su_X_test, su_y_test)
```

### Classification

```python
HN_FAMOUS_THRESHOLD = 2500
SU_FAMOUS_THRESHOLD = 1000
```

```python
hn_is_famous_train = hn_y_train > HN_FAMOUS_THRESHOLD
su_is_famous_train = su_y_train > SU_FAMOUS_THRESHOLD
print "HN number of famous nodes (train): %d" % np.sum(hn_is_famous_train)
print "HN number of non-famous nodes (train): %d" % np.sum(np.invert(hn_is_famous_train))
print "SU number of famous nodes (train): %d" % np.sum(su_is_famous_train)
print "SU number of non-famous nodes (train): %d" % np.sum(np.invert(su_is_famous_train))
```

```python
hn_is_famous_test = hn_y_test > HN_FAMOUS_THRESHOLD
su_is_famous_test = su_y_test > SU_FAMOUS_THRESHOLD
print "HN number of famous nodes (test): %d" % np.sum(hn_is_famous_test)
print "HN number of non-famous nodes (test): %d" % np.sum(np.invert(hn_is_famous_test))
print "SU number of famous nodes (test): %d" % np.sum(su_is_famous_test)
print "SU number of non-famous nodes (test): %d" % np.sum(np.invert(su_is_famous_test))
```

```python
hn_logistic = linear_model.LogisticRegression(C=1e5)
hn_probs = hn_logistic.fit(hn_X_train, hn_is_famous_train).predict_proba(hn_X_test)
```

```python
su_logistic = linear_model.LogisticRegression(C=1e5)
su_probs = su_logistic.fit(su_X_train, su_is_famous_train).predict_proba(su_X_test)
```

```python
hn_fpr, hn_tpr, _ = roc_curve(hn_is_famous_test, hn_probs[:, 1])
su_fpr, su_tpr, _ = roc_curve(su_is_famous_test, su_probs[:, 1])
hn_roc_auc = auc(hn_fpr, hn_tpr)
su_roc_auc = auc(su_fpr, su_tpr)

plt.figure(figsize=(8,8))
plt.plot(hn_fpr, hn_tpr, label='Hacker News ROC curve (area = %0.2f)' % hn_roc_auc)
plt.plot(su_fpr, su_tpr, label='Super User ROC curve (area = %0.2f)' % su_roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.title("Receiver Operating Characteristic")
plt.legend(loc="lower right")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.savefig(os.path.join(FIGURES_DIR, "classification_roc.pdf"))
```

```python
hn_precision, hn_recall, _ = precision_recall_curve(hn_is_famous_test, hn_probs[:, 1])
su_precision, su_recall, _ = precision_recall_curve(su_is_famous_test, su_probs[:, 1])

plt.figure(figsize=(8,8))
plt.plot(hn_recall, hn_precision, label='Hacker News Precision-Recall curve')
plt.plot(su_recall, su_precision, label='Super User Precision-Recall curve')
plt.plot([0, 1], [1, 0], 'k--')
plt.title('Precision-Recall')
plt.legend(loc="lower left")
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.savefig(os.path.join(FIGURES_DIR, "classification_pr_curve.pdf"))
```
